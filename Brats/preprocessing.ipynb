```python
import scipy
from scipy.ndimage import zoom
import nibabel as nib
import matplotlib.pyplot as plt
import numpy as np
```

```python
def save_resized(save_dir,file_path):
    x = nib.load(file_path).get_fdata()
    zoom_factors = (128/240, 128/240, 128/155)  
    resized_x = zoom(x, zoom_factors, order=1)
   
    resized_x = (resized_x).astype(np.float32)
    resized_img = nib.Nifti1Image(resized_x, np.eye(4))
    save_path = os.path.join(save_dir,"_128_" + file_path.split('\\')[-1])
    print("saving at",save_path)
    nib.save(resized_img, save_path)
```

```python
import os
base_dir = None 
save_seg,save_img = None,None
# save_img and save_seg will separately store segmentation masks and input mri's after resizing them.
for patients in os.listdir(base_dir):
    patient_path = os.path.join(base_dir,patients)
    for file in os.listdir(patient_path):
        file_path = os.path.join(patient_path,file)
        print(file_path)
        if("Binary" in file_path):
            save_resized(save_seg,file_path)
        elif("t1ce" in file_path):
            save_resized(save_img,file_path)
        
```

```python
import os
import nibabel as nib
import numpy as np
images=[]
img_dir = None # dir that has all the 3d input volumes to be converted into a single npy array to directly import in kaggle
for img in os.listdir(img_dir):
    img_path = os.path.join(img_dir,img)
    arr = nib.load(img_path).get_fdata().astype(np.float32)
    arr = (arr-arr.mean())/arr.std()
    images.append(arr)

seg_dir = None # dir that has all the 3d segmentation masks to be converted into a single npy array to directly import in kaggle

segs=[]
for img in os.listdir(seg_dir):
    seg_path = os.path.join(seg_dir,img)
    arr = nib.load(seg_path).get_fdata().astype(np.float32)
    segs.append(arr)
save_path = None
np.save(os.path.join(save_path,"t1_ce_scans_resized.npy"),np.stack(images))
np.save(os.path.join(save_path,"seg_masks_resized.npy"),np.stack(segs))


    
```
